```python
import re

def extract_unit_price_usd(description):
    if not isinstance(description, str):
        return None
    match = re.search(r'USD\s*([0-9.]+)', description.upper())
    if match:
        return float(match.group(1))
    return None

df['EXTRACTED_UNIT_PRICE_USD'] = df['GOODS DESCRIPTION'].apply(extract_unit_price_usd)

df[['GOODS DESCRIPTION', 'EXTRACTED_UNIT_PRICE_USD']].head(10)
```

```python
def extract_embedded_quantity(description):
    if not isinstance(description, str):
        return None
    # Match patterns like "QTY:600", "QTY 6336", or "QTY: 336000"
    match = re.search(r'QTY[\s:]*([0-9,]+)', description.upper())
    if match:
        num = match.group(1).replace(',', '')
        return int(num)
    return None

df['EXTRACTED_QUANTITY'] = df['GOODS DESCRIPTION'].apply(extract_embedded_quantity)
df[['GOODS DESCRIPTION', 'EXTRACTED_QUANTITY']].head(10)
```

```python
import re

def extract_unit_price_usd(description):
    if not isinstance(description, str):
        return None
    match = re.search(r'USD\s*([0-9.]+)', description.upper())
    if match:
        return float(match.group(1))
    return None

df['EXTRACTED_UNIT_PRICE_USD'] = df['GOODS DESCRIPTION'].apply(extract_unit_price_usd)
df[['GOODS DESCRIPTION', 'EXTRACTED_UNIT_PRICE_USD']].head(10)
```

```python
def extract_model_name(description):
    if not isinstance(description, str):
        return None
    # Tries to match model names like "XXXX", "XXXX-XX", "SB-12", etc. at start of string
    match = re.match(r'([A-Z0-9-]+)', description.strip())
    if match:
        return match.group(1)
    return None

df['EXTRACTED_MODEL_NAME'] = df['GOODS DESCRIPTION'].apply(extract_model_name)
df[['GOODS DESCRIPTION', 'EXTRACTED_MODEL_NAME']].head(10)
```

```python
def extract_model_name(description):
    if not isinstance(description, str):
        return None
    # Find code: starts with letters, then digits and optionally dashes, e.g. "TH5170", "8001-2", "AM-967"
    match = re.match(r'([A-Z]+[-]?[0-9]+(?:-[0-9]+)?)', description.strip())
    # If not found, tries pattern: parentheses like "(RYX-02..." or "(SL-89)"
    if not match:
        match = re.search(r'\(([A-Z0-9-]+)\)', description)
    if match:
        return match.group(1)
    return None

df['EXTRACTED_MODEL_NAME'] = df['GOODS DESCRIPTION'].apply(extract_model_name)
df[['GOODS DESCRIPTION', 'EXTRACTED_MODEL_NAME']].head(10)
```

```python
def extract_capacity(description):
    if not isinstance(description, str):
        return None
    # Find typical capacity patterns: e.g., "750ML", "2PCS", "10SET", "48 LTR", "22MM"
    match = re.search(r'([0-9,.]+)\s?(ml|ltr|kg|set|pcs|mm|cm|mt)', description.lower())
    if match:
        return match.group(0)
    return None

df['EXTRACTED_CAPACITY'] = df['GOODS DESCRIPTION'].apply(extract_capacity)
df[['GOODS DESCRIPTION', 'EXTRACTED_CAPACITY']].head(10)
```

```python
def extract_material_type(description):
    if not isinstance(description, str):
        return None
    # Match common material types
    materials = ['steel', 'plastic', 'glass', 'aluminum', 'wood', 'polyhouse', 'borosilicate']
    for m in materials:
        if m in description.lower():
            return m.title()
    return None

df['EXTRACTED_MATERIAL'] = df['GOODS DESCRIPTION'].apply(extract_material_type)
df[['GOODS DESCRIPTION', 'EXTRACTED_MATERIAL']].head(10)
```

```python
def extract_model_number(description):
    if not isinstance(description, str):
        return None
    # First, look for a code inside parentheses (e.g., "(SL-89)", "(RYX-02-020)")
    match = re.search(r'\(([A-Z0-9\-]+)\)', description)
    if match:
        return match.group(1)
    # If not found, look for code at the start (if description starts with a code)
    match = re.match(r'([A-Z0-9\-]+)', description.strip())
    if match:
        return match.group(1)
    return None

df['EXTRACTED_MODEL_NUMBER'] = df['GOODS DESCRIPTION'].apply(extract_model_number)
df[['GOODS DESCRIPTION', 'EXTRACTED_MODEL_NUMBER']].head(10)
```

```python
unit_map = {
    'nos': 'pcs',
    'pieces': 'pcs',
    'pcs': 'pcs',
    'kgs': 'kg',
    'kg': 'kg',
    'mt': 'mt',
    'sets': 'set',
    'set': 'set'
}

df['UNIT_STD'] = df['UNIT'].str.lower().map(unit_map).fillna(df['UNIT'].str.lower())
df[['UNIT', 'UNIT_STD']].head(10)
```

```python
df['LANDED_COST'] = df['TOTAL VALUE_INR'] + df['DUTY PAID_INR']
df[['TOTAL VALUE_INR', 'DUTY PAID_INR', 'LANDED_COST']].head(10)
```

```python
def assign_category(desc, hsn):
    if not isinstance(desc, str): return 'Other'
    d = desc.lower()
    hsn = str(hsn)
    # Steel (general, by keywords or HSN)
    if "steel" in d or hsn.startswith("73"):
        return "Steel"
    if "glass" in d or hsn.startswith("70"):
        return "Glass"
    if "plastic" in d or hsn.startswith("39"):
        return "Plastic"
    if "polyhouse" in d:
        return "Polyhouse"
    if "wood" in d or hsn.startswith("44"):
        return "Wooden"
    return "Other"

df['CATEGORY'] = df.apply(lambda x: assign_category(x['GOODS DESCRIPTION'], x['HS CODE']), axis=1)
```

```python
def assign_subcategory(desc):
    if not isinstance(desc, str): return 'Other'
    d = desc.lower()
    if "scrubber" in d:     return "Scrubber"
    if "basket" in d:       return "Basket"
    if "strainer" in d:     return "Strainer"
    if "hanger" in d:       return "Hanger"
    if "cutlery" in d:      return "Cutlery"
    if "bottle" in d:       return "Bottle"
    if "stand" in d:        return "Stand"
    if "fridge" in d:       return "Fridge Item"
    return "Other"

df['SUB_CATEGORY'] = df['GOODS DESCRIPTION'].apply(assign_subcategory)
df[['GOODS DESCRIPTION', 'CATEGORY', 'SUB_CATEGORY']].head(10)
```

```python
import pandas as pd

df_macro = pd.read_sql_query(
    "SELECT YEAR, SUM([TOTAL VALUE_INR]) AS total_imports, SUM([DUTY PAID_INR]) AS duty_paid \
     FROM cleaned_trade_data GROUP BY YEAR ORDER BY YEAR", conn
)
print(df_macro)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate and prepare your growth data
df_macro['YoY_imports'] = df_macro['total_imports'].pct_change() * 100
df_macro['YoY_duty'] = df_macro['duty_paid'].pct_change() * 100

heatmap_data = df_macro[['YEAR', 'YoY_imports', 'YoY_duty']].set_index('YEAR')

plt.figure(figsize=(10, 3))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".1f",
    cmap='Spectral',
    linewidths=2,
    linecolor='white',
    cbar_kws={'label': 'YoY Growth %'}
)
plt.title('Year-over-Year Growth (%) of Total Imports and Duty Paid')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
